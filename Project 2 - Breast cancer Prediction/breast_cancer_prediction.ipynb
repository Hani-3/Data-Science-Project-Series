```python
from sklearn.datasets import load_breast_cancer
```

```python
#load dataset
data = load_breast_cancer()
```

```python
# Split dataset in X and Y
import pandas as pd
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target)
```

```python
import matplotlib.pyplot as plt
plt.hist(y)
plt.title('Diagnosis [malignant = 1, benign = 0]')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(20,15))
graph=1
for column in X:
    ax = plt.subplot(5,6, graph)
    sns.histplot(X[column])
    plt.xlabel(column)
    graph+=1

plt.tight_layout()
plt.show()
```

```python
# heatmap to see correlation between data
plt.figure(figsize=(20,12))
sns.heatmap(X.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#scaling
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
svc = SVC()
param_grid = {
    'C':[0.01,0.1,10,50,100,1000],
    'gamma':[0.001,0.01,0.1,1],
    'kernel': ['linear','rbf','poly']
}
grid_search = GridSearchCV(svc,param_grid=param_grid,refit=True,cv=5,verbose=3)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
svc = SVC(C=0.1,gamma=0.001,kernel='linear')
svc.fit(X_train,y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print('accuracy score :', accuracy_score(y_test, y_pred))
print('confusion matrix: \n', confusion_matrix(y_test, y_pred))
print('classsification report: \n', classification_report(y_test, y_pred))
```
